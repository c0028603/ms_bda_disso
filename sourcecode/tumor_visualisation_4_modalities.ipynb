```python
import pandas as pd
import numpy as np
from scipy.sparse import coo_matrix
from matplotlib import pyplot as plt
```

```python
def get_csv_df_to_img(csv_df):
    sparse_image = coo_matrix((csv_df.VoxVal, (csv_df.X,csv_df.pY)), shape = (240,240))
    image = sparse_image.todense()
    return image 

```

```python
flair_gray_df = pd.read_csv('C:/Drive/Dissertation/OneDrive_1_23-06-2022/BraTSReg_gray_matter_csv_v2/BraTSReg_001_00_0000_flair.nii.gz_gray_matter.csv')
t1ce_gray_df = pd.read_csv('C:/Drive/Dissertation/OneDrive_1_23-06-2022/BraTSReg_gray_matter_csv_v2/BraTSReg_001_00_0000_t1ce.nii.gz_gray_matter.csv')
t1ce_full_df = pd.read_csv('C:/Drive/Dissertation/OneDrive_1_23-06-2022/BraTSReg_Training_Data_v2_csv_4_modality/BraTSReg_001_00_0000_t1ce_values.csv')
flair_full_df = pd.read_csv('C:/Drive/Dissertation/OneDrive_1_23-06-2022/BraTSReg_Training_Data_v2_csv_4_modality/BraTSReg_001_00_0000_flai_values.csv')
slice = 51
z_flair_gray = flair_gray_df.query('Z == 70')
z_t1ce_gray = t1ce_gray_df.query('Z == 70')
z_flair_full = flair_full_df.query('Z == 70')
z_t1ce_full = t1ce_full_df.query('Z == 70')
image_flair_gray = get_csv_df_to_img(z_flair_gray)
image_t1ce_gray = get_csv_df_to_img(z_t1ce_gray)
image_flair_full = get_csv_df_to_img(z_flair_full)
image_t1ce_full = get_csv_df_to_img(z_t1ce_full)

img_lst = [image_flair_gray[141:161,131:151],image_t1ce_gray[141:161,131:151],image_flair_full[121:181,111:171],image_t1ce_full[121:181,111:171],image_flair_full,image_t1ce_full]

```

```python
w = 50
h = 30
fig = plt.figure(figsize=(10, 10))
columns = 2
rows = 3
for i in range(1, columns*rows +1):
    #img = np.random.randint(10, size=(h,w))
    ax = fig.add_subplot(rows, columns, i)
    if (i % 2) == 0:
       ax.title.set_text('t1ce Image ' )
    else:
        ax.title.set_text('Flair Image ')
    plt.gca().invert_yaxis()
    plt.imshow(img_lst[i-1],cmap='gray')
plt.show()
```
